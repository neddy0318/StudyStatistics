# 분산분석 ANOVA; Analysis of Variance

## 1. 개요

- ANOVA는 1918 로날드 피셔(Ronal Fisher)가 고안한 통계학적 테스트입니다.
- 세 개 이상 독립적인 그룹의 평균 사이에 어떤 통계학적 차이가 있는지를 알아보는 것이 주된 목적입니다.
- F분포를 사용합니다. F검정의 통계갓은 집단 간 분산의 비율을 나타내며, 그렇기에 분산분석이라고 불립니다.
- *t-test : T분포를 활용하는 t검정은 두 집단 간 평균 차이를 검정하며, 이를 통해 집단 A와 B를 비교하고 A와 C를 비교하는 방식으로 세 집단 간 평균 차이를 검정할 수 있습니다. 하지만 이럴 경우 신뢰도가 하락하기에 세 집단 이상의 비교에는 F검정을 활용합니다.


## 2. F-분포와 ANOVA
- 연속활률분포로, 두 모분산의 비교 및 검정에 사용합니다.
- 항상 양의 값을 가지며, 오른쪽으로 긴 꼬리 형태를 가집니다.
- ANOVA나 회귀분석등에 활용되며, 독립변수가 종속변수에 미치는 영향을 검정하는 것이 주된 목적입니다.
- ANOVA도 결국 집단의 종류(독립변수)가 평균 값의 차이 여부(종속변수)에 미치는 영향을 검정하는 것이기에, 회귀분석과 비슷한 측면이 있습니다.


## 3. 기본 개념과 가설
1)기본 개념
- ANOVA의 기본 아이디어는 집단 간 분산(between-group variance)과 집단 내 분산(within-group variance)을 비교하는 것입니다.
    - 집단 간 분산 : 집단 평균들 사이의 차이에 의한 분산. 각 집단의 평균이 얼마나 다른지를 측정합니다.
    - 집단 내 분산 : 각 집단 내부에서 개별 관측값들이 그 집단의 평균으로부터 얼마나 떨어져 있는지에 따른 분산. 이는 집단 내의 일관성을 의미합니다.
- 두 분산을 비교하여, 집단 간 분산이 집단 내 분산보다 크다면 집단 간 평균이 다르다고 결론 내릴 수 있습니다.
2)기본 가설
- 위의 기본적인 개념에 따라, ANOVa의 일반적인 가설은 다음과 같은 형태가 됩니다.
    - H0(귀무가설) : 독립변수(인자)의 차이에 따른 종속변수(특성값)은 동일하다.
    - H1(대립가설) : 독립변수(인자)의 차이에 따른 종속변수(특성값)은 다르다.
- 독립변수인 요인의 수에 따라 ANOVA의 종류가 달라지는데, 이는 아래에서 다시 살펴보도록 하겠습니다.
- #참고 : 독립변수는 범주(분류)형 변수여야하며, 종속 변수는 연속형 변수여야합니다.


## 4. 기본 가정
- 정규성 가정: 각 집단의 데이터가 정규분포를 따른다고 가정합니다.
- 등분산성 가정: 각 집단의 분산이 같아야 합니다.
- 독립성 가정: 각 집단의 관측값은 서로 독립적이어야 합니다.


## 5. ANOVA의 종류
ANOVA는 비교하고자 하는 요인(factor)의 수에 따라 여러 가지 유형이 있습니다.

- **일원 분산분석(One-Way ANOVA)**: 하나의 요인에 대해 두 개 이상의 집단을 비교할 때 사용됩니다.  
  예: 세 그룹(A, B, C) 간 평균 키의 차이를 비교하는 경우.

- **이원 분산분석(Two-Way ANOVA)**: 두 개의 요인을 고려하여 집단 간 차이를 분석합니다. 두 요인의 상호작용(interaction)도 분석할 수 있습니다.  
  예: 교육 수준과 나이대가 시험 성적에 미치는 영향을 동시에 비교할 때.

- **반복 측정 분산분석(Repeated Measures ANOVA)**: 같은 집단에 대해 여러 시점에서 측정된 데이터에 사용됩니다.  
  예: 운동 프로그램 전, 중, 후의 체중 변화를 비교할 때.


#### 5-1. 일원 분산분석 예
간단한 **일원 분산분석(One-Way ANOVA)** 절차는 다음과 같습니다:

- 귀무가설(H₀)과 대립가설(H₁)

  - **귀무가설(H₀)**: 모든 집단의 평균이 같다.
  $$ H_0: \mu_1 = \mu_2 = \mu_3 = \dots = \mu_k $$

  - **대립가설(H₁)**: 적어도 하나의 집단 평균이 다르다.
  $$ H_1: \text{At least one group mean is different} $$

#### 5-2. 분산분석의 기본 개념

ANOVA는 총 변동을 **집단 간 변동**과 **집단 내 변동**으로 나눕니다.

- **총 제곱합 (SST)**: 전체 데이터의 총 변동을 나타냅니다.
  $$ SST = \sum_{i=1}^{N} (Y_i - \bar{Y})^2 $$
  여기서 \( Y_i \)는 각 관측치, \( \bar{Y} \)는 전체 평균, \( N \)은 총 샘플 수입니다.

- **집단 간 제곱합 (SSB)**: 집단 간 변동을 나타내며, 각 그룹 평균과 전체 평균 간의 차이를 나타냅니다.
  $$ SSB = \sum_{j=1}^{k} n_j (\bar{Y}_j - \bar{Y})^2 $$
  여기서 \( n_j \)는 그룹 \( j \)의 샘플 크기, \( \bar{Y}_j \)는 그룹 \( j \)의 평균입니다.

- **집단 내 제곱합 (SSW)**: 집단 내 변동을 나타내며, 그룹 내 관측값들이 그룹 평균으로부터 얼마나 벗어나는지 나타냅니다.
  $$ SSW = \sum_{j=1}^{k} \sum_{i=1}^{n_j} (Y_{ij} - \bar{Y}_j)^2 $$

#### 5-3. F-통계량 계산

F-통계량은 집단 간 변동(SSB)을 집단 내 변동(SSW)으로 나눈 값으로 계산됩니다. 이를 통해, 집단 간 차이가 통계적으로 유의미한지 평가할 수 있습니다.

- **집단 간 평균 제곱 (MSB)**: 집단 간 변동을 자유도(df)로 나눈 값입니다.
  $$ MSB = \frac{SSB}{k-1} $$

- **집단 내 평균 제곱 (MSW)**: 집단 내 변동을 자유도(df)로 나눈 값입니다.
  $$ MSW = \frac{SSW}{N - k} $$

- **F-통계량**: 집단 간 변동과 집단 내 변동의 비율로 계산됩니다.
  $$ F = \frac{MSB}{MSW} $$

#### p-value 계산

- F-값에 대응하는 **p-value**를 계산하고, p-value가 유의 수준(α)보다 작으면 귀무가설을 기각합니다.

### 6. ANOVA 결과 해석

분산분석의 결과는 일반적으로 **ANOVA Table**이라는 표로 요약됩니다.

| Source          | SS (Sum of Squares) | df  | MS (Mean Square)  | F       | p-value |
|-----------------|---------------------|-----|-------------------|---------|---------|
| Between Groups  | \( SSB \)            | \( k-1 \) | \( MSB = \frac{SSB}{k-1} \) | \( F = \frac{MSB}{MSW} \) | p-value |
| Within Groups   | \( SSW \)            | \( N-k \) | \( MSW = \frac{SSW}{N-k} \) |         |         |
| **Total**       | \( SST \)            | \( N-1 \) |                   |         |         |

- **F-통계량**: 집단 간 평균 차이의 정도를 나타냅니다.
- **p-value**: 집단 간 차이가 우연히 발생할 확률을 나타냅니다. 일반적으로 유의 수준(α)을 0.05로 설정합니다. p-value가 0.05 미만이면 귀무가설을 기각하고, 집단 간 평균이 통계적으로 유의미하게 다르다고 결론지을 수 있습니다.

### 6. 사후검정(Post-hoc Test)

ANOVA는 집단 간 평균 차이가 있는지를 확인하는 데 중점을 두지만, **어느 집단끼리 차이가 있는지**는 알 수 없습니다. 이를 위해 **사후검정**을 수행합니다. 대표적인 사후검정 방법은 **Tukey's HSD 검정**입니다.

### 예시

어떤 실험에서 3개의 다른 약물을 테스트한 후, 각 약물에 대한 환자의 혈압 변화량을 측정했다고 가정합니다. 이 경우, **일원 분산분석**을 통해 약물 간 혈압 변화량에 차이가 있는지 분석할 수 있습니다.

- **귀무가설**: 세 그룹(약물 A, B, C) 간 평균 혈압 변화량이 같다.
- **대립가설**: 적어도 하나의 약물 그룹이 다른 그룹과 평균 혈압 변화량이 다르다.

분산분석을 통해 F-값이 계산되고, p-value가 0.03이라면 이는 유의 수준 0.05보다 작으므로 귀무가설을 기각하고, 약물 간 혈압 변화량이 유의미하게 다르다고 결론을 내립니다.

## 파이썬 코드 

*Chat-gpt 활용

```python
import numpy as np
from scipy import stats

# 예제 데이터: 3개의 약물 그룹 A, B, C의 혈압 변화량
group_A = [5.2, 5.7, 5.1, 5.6, 5.3]
group_B = [4.8, 4.9, 5.0, 4.7, 5.1]
group_C = [6.1, 6.2, 6.3, 6.0, 6.4]

# 일원 분산분석 수행
f_statistic, p_value = stats.f_oneway(group_A, group_B, group_C)

print(f"F-Statistic: {f_statistic}")
print(f"P-Value: {p_value}")

# 결과 해석
alpha = 0.05
if p_value < alpha:
    print("귀무가설을 기각합니다. 집단 간 평균에 유의미한 차이가 있습니다.")
else:
    print("귀무가설을 기각하지 않습니다. 집단 간 평균에 유의미한 차이가 없습니다.")
```
