# p 값과 통계적 유의성 P-value and Statistic Significance

- 지난번에는 가설검정에 대해 살펴보면서 같이 p 값과 통계적 유의성도 다뤘습니다. 이번에는 둘에 대해 좀 더 자세히 살펴보도록 하겠습니다.

## 1. p값 개요

### 1)개요
- p값은 귀무가설이 우연히 성립될 확률입니다.
- 또는 귀무가설(null hypothesis)이 참이라는 가정 하에, 관찰된 데이터(또는 더 극단적인 결과)가 나타날 확률을 나타내는 값입니다. 
- 가령 p값이 0.01이라는 것은 귀무가설이 우연히 성립하는 경우가 100번 중 1회(1%) 있다는 의미입니다.
    - p값이 작을수록 현재 데이터가 귀무가설 하에서 나타날 가능성이 적다는 것을 의미합니다.
    - p값이 클수록 현재 데이터가 귀무가설 하에서 나타날 가능성이 높다는 것을 의미합니다.

### 2)p값 해석 시 주의할 점 및 참고

- (1)주의할 점
    - 앞서 정리한 바와 같이, 통상 p값을 통상 '귀무가설을 참이라고 가정할 때, 관측한 데이터의 통계량이 귀무가설을 지지할 확률', 혹은 '귀무가설이 우연히 성립할 확률' 등으로 정의햇지만, 미국 통계학회(ASA)는 이러한 정의가 꼭 옳은 것은 아니라고 지적했습니다.
    - 즉, p값이 어떤 가설의 참 거짓 여부, 혹은 실험 결과의 중요성 여부를 결정할 수 없다는 것입니다.
    - ASA는 p값이 0.05보다 작다는 것을 '주어진 가설이 참일 확률이 0.95% 이상'이라고 해석하기보다, 귀무가설과 다른 모든 가설이 타당하다는 전제 하에서, 관찰된 데이터의 결과값이 그 이상으로 극단적인 결과를 얻을 확률이 5%보다 낮다'라고 해석하는 것이 맞다고 주장하는 것입니다.
    - 다른 식으로 이야기하면, p값은 내가 관찰한 데이터 값이 귀무가설의 통계 모형과 얼마나 상반되는지 (귀무가설의 확률 분포에서 이러한 실험 데이터가 나올 확률이 얼마나 낮은지)를 나타내는 값입니다.
- (2)정리
    - p값이 통계적 유의성 확보를 위한 도구이나, 실험의 중요성이나 가설이 참임을 주장하기 위한 절대적인 지표는 아닙니다.
    - 올바른 적용과 해석을 위해서는 유의수준과 실험 규모등 다른 요소들을 적절히 고려해야합니다.


## 2. 유의수준과 통계적 유의성

- 유의수준과 통계적 유의성은 비슷하고 서로 관련된 개념이지만, 그 차이점과 관계를 이해하고 구별해야 합니다.
- 결론부터 이야기하면, `유의수준은 판단의 기준이고, 통계적 유의성은 그에 따른 결과입니다.`

### 1)유의수준(α)
- (1)개요
    - 유의수준은 연구자가 가설 검정에서 통계적 유의성을 판단하기 위해 설정한 기준입니다. 
    - p값과 비교하여 귀무가설을 기각할지 여부를 결정하는 임계값입니다. 일반적으로 0.05(5%) 또는 0.01(1%)이 자주 사용됩니다.
    - 1종 오류를 저지를 확률이기도 합니다. 예를 들어, 유의수준을 0.05로 설정했다면, 귀무가설이 참일 때 이를 기각할 확률이 5%라는 의미입니다.
    - 유의수준이 0.05라면, p값이 0.05 이하일 때 귀무가설을 기각하고 결과를 통계적으로 유의미하다고 판단합니다.
    - 반대로 p값이 0.05보다 크다면 귀무가설을 기각하지 않고, 결과를 우연에 의해 발생했을 가능성이 크다고 판단합니다.
- (2)p값과의 관계
    - p값 ≤ α: 귀무가설을 기각합니다. 즉, 결과는 통계적으로 유의미합니다. 관찰된 데이터가 우연이 아닌 실제 차이나 효과를 나타낼 가능성이 높다고 판단합니다.
    - p값 > α: 귀무가설을 기각하지 않습니다. 즉, 결과는 통계적으로 유의미하지 않습니다. 관찰된 데이터가 우연에 의한 결과일 가능성이 높다고 판단합니다.

### 2)통계적 유의성
- (1)개요
    - 통계적 유의성은 어떤 연구 결과가 우연히 발생한 것이 아니라, 실제로 의미 있는 차이나 효과가 있다고 판단할 수 있는지 여부를 말합니다. 즉, 데이터 분석 결과가 통계적으로 유의미한지를 결정하는 것입니다.
    - 특정 결과가 통계적으로 유의미하다는 것은, 우리가 설정한 유의수준(보통 5% 또는 0.05)보다 작은 p값을 얻었을 때 귀무가설을 기각하고 대립가설을 채택할 수 있음을 의미합니다. 이는 결과가 우연히 발생할 확률이 매우 낮음을 나타냅니다.
- (2)정리 및 유의수준과의 관계
    - 유의수준(α)은 연구자가 미리 설정하는 값으로, p값과 비교하여 통계적 유의성을 판단하는 기준 역할을 합니다.
    - 통계적 유의성은 이 기준을 바탕으로 한 결과로, 실제로 통계적으로 유의미한 차이나 효과가 있다고 결론짓는 것입니다.
    - #참고
        - 귀무가설이 맞는데 귀무가설을 기각하는 것은 1종 오류입니다.
        - 귀무가설이 거짓인데도 불구하고 이를 기각하지 않는 오류입니다.



## 3. p값 계산과 예시 및 코드 -- 챗gpt 활용

#### 1. p값(p-value) 계산 공식

p값은 주로 표본 데이터에서 관찰된 통계량이 귀무가설 하에서 얼마나 극단적인지를 나타내는 확률로 계산됩니다. p값을 구하는 방법은 통계 검정 방법에 따라 달라지지만, 일반적인 절차는 다음과 같습니다:

1. **통계량 계산**: 관찰된 데이터에 대해 해당 검정에 맞는 통계량(예: z값, t값)을 계산합니다.

   예를 들어, z검정의 경우:

   $$
   z = \frac{X̄ - μ0}{σ / √n}
   $$

   여기서 X̄는 표본 평균, μ0는 귀무가설 하의 모평균, σ는 모표준편차, n은 표본 크기입니다.

2. **분포에서 p값 추정**: 해당 통계량을 바탕으로 정규 분포(또는 t분포 등)에서 p값을 계산합니다.

   p값은 검정 통계량이 실제로 관측된 것보다 더 극단적이거나 같은 값을 가질 확률입니다.

   - 단측 검정:

     $$
     p = P(Z ≥ |z|)
     $$

   - 양측 검정:

     $$
     p = 2 × P(Z ≥ |z|)
     $$

3. **p값 비교**: 계산된 p값을 유의수준 α와 비교하여 귀무가설을 기각할지 여부를 결정합니다.

#### 2. 예시

##### 예시 시나리오:

한 연구자가 새로 개발된 약물이 기존 약물보다 평균적으로 더 높은 효과가 있는지 검정하려고 합니다. 기존 약물의 평균 효과는 50이며, 새 약물의 표본 평균 효과는 55, 표준편차는 10, 표본 크기는 30입니다. 연구자는 유의수준 α = 0.05에서 이 가설을 검정합니다.

1. **귀무가설과 대립가설 설정**

   - 귀무가설 (H0): μ = 50 (새 약물의 평균 효과가 기존 약물과 같다)
   - 대립가설 (H1): μ > 50 (새 약물의 평균 효과가 기존 약물보다 크다)

2. **z값 계산**

   $$
   z = \frac{X̄ - μ0}{σ / √n} = \frac{55 - 50}{10 / √30} ≈ 2.74
   $$

3. **p값 계산**

   단측 검정에서 z값이 2.74인 경우의 p값은 정규 분포표에서 구할 수 있습니다.

   $$
   P(Z ≥ 2.74)
   $$

   를 계산하면 p값은 약 0.0031입니다.

4. **p값과 유의수준 비교**

   p값 = 0.0031은 유의수준 α = 0.05보다 작으므로, 귀무가설을 기각할 수 있습니다.

5. **결론**

   새 약물이 기존 약물보다 유의미하게 더 높은 효과가 있다는 결론을 내릴 수 있습니다.

```python
import numpy as np
from scipy import stats

# 데이터 설정
sample_mean = 55        # 새 약물의 표본 평균
population_mean = 50    # 기존 약물의 모평균
std_dev = 10            # 표준편차
sample_size = 30        # 표본 크기

# z-값 계산
z_value = (sample_mean - population_mean) / (std_dev / np.sqrt(sample_size))

# p값 계산 (단측 검정)
p_value = 1 - stats.norm.cdf(z_value)

print(f"z-value: {z_value:.2f}")
print(f"p-value: {p_value:.4f}")

# 유의수준 설정
alpha = 0.05

# 귀무가설 기각 여부 결정
if p_value < alpha:
    print("귀무가설을 기각합니다. 새 약물의 평균 효과가 기존 약물보다 유의미하게 높습니다.")
else:
    print("귀무가설을 기각하지 않습니다. 새 약물의 평균 효과가 기존 약물과 차이가 없습니다.")
```
