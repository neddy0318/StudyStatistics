# 중심극한 정리 CLT; Central Limit Theorem

## 1. 개요



*출처 : https://www.kaggle.com/code/carlmcbrideellis/animated-histogram-of-the-central-limit-theorem*


#### 1)정의
- "많은 경우 독립적이고 동일한 분포(iid; independent and identically distributed)를 갖는 확률변수의 경우 원래 변수 자체가 정규 분포를 따르지 않더라도 표본 평균의 분포가 정규 분포를 따르는 경향이 있다."
- 모집단에서 표본크기가 n인 표본을 여러번 반복해서 추출했을 때 각각의 표본 평균들이 이루는 분포는 정규분포에 가깝다(단, n이 충분히 커야함).
- 원래 데이터의 분포가 무엇이든 표본 평균의 분포가 정규 분포를 따른다.

#### 2)수학적 정의 및 표준화

**(1)표본 평균의 분포**

먼저, 중심극한정리에 따르면 표본 평균이 충분히 큰 표본 크기에서 정규분포에 수렴합니다. 여기서 중요한 개념은 다음과 같습니다:

- **표본 평균** X̄_n: 주어진 모집단에서 n개의 표본을 추출하고 그 평균을 계산한 것입니다.
- **모집단의 평균** μ: 표본이 추출된 모집단의 평균입니다.
- **모집단의 분산** σ²: 표본이 추출된 모집단의 분산입니다.

**(2)표준화되지 않은 표본 평균의 분포**

표본 크기 n이 충분히 클 때, 표본 평균 X̄_n은 다음과 같은 정규분포를 따릅니다:

$$
X̄_n ∼ N(μ, σ² / n)
$$

여기서:
- μ는 모집단 평균입니다.
- σ² / n은 표본 평균의 분산입니다.

이 의미는, 표본 평균 X̄_n이 평균이 μ이고, 분산이 σ² / n인 정규분포에 가까워진다는 것입니다.

**(3)표준화된 표본 평균의 분포**

- 표본 평균을 표준화한다는 것은, 표본 평균에서 모집단의 평균 μ를 빼고, 이를 표본 평균의 표준편차 σ / √n으로 나누는 것입니다.
- 표준 정규분포를 이용하면, 다양한 정규분포의 값을 공통된 기준으로 비교하고 계산할 수 있기 때문에, 통계적 분석에서 매우 중요한 역할을 합니다. 이 분포는 통계적 추정과 가설 검정에서 널리 사용됩니다. 

$$
Z = (X̄_n - μ) / (σ / √n)
$$

이때 Z는 표준 정규분포 N(0, 1)을 따릅니다.


#### 3)주의
잘못된 해석들
- 표본의 분포가 아닌, 표분(표본 분포)의 평균입니다!
- 표본의 양이 충분하면 모집단의 평균과 유사해진다는 뜻이 아닙니다.
    - #참고 : 대수의 법칙(=큰수의 법칙, LLN; Law of Large Numbers)
        - 표본 평균은 표본의 크기(n)가 커질수록 모집단의 평균에 가까워집니다.
        - 큰수의 법칙이 표본 평균이 모집단의 평균으로 수렴함을 보장해주는 반면, 중심극한정리는 이 과정에서 표본 평균이 어떤 분포를 따르게 되는지 설명해줍니다. 이는 특히 표본이 큰 경우에 통계적 분석을 수행할 때 중요한 역할을 합니다.

#### 4)중요한 이유
- 모집단의 분포와 상관 없이 표분 평균이 정규분포를 이룬다는 점을 이용하면 특정 사건이 일어날 확률을 계산할 수 있게 됩니다. 즉, 표본 평균들이 이루는 표분과 모집단 간 관계를 증명함으로써 모집단의 모수를 추정할 수 있는 수학적(확률적) 근거를 마련해줍니다.
- 모집단의 분포에 대한 정보가 아예 없더라도, 표본 통계량을 이용해 모수를 추정할 수 있는 강력한 이론적 증거를 제시한다.
- 표준정규분포표가 주어지면 확률값을 구하기 쉽습니다.

## 2. 조건
앞서 '독립적이고 동일한 분포(idd)를 갖는 확률변수'를 언급했듯, 중심극한정리가 성립하기 위해서는 다음의 조건이 전제되어야합니다.
- **표본 크기:** 𝑛이 충분히 커야 합니다. 일반적으로 n≥30이 권장됩니다.
- **독립성:** 표본은 서로 독립적이어야 합니다. 즉, 개별 관측치 사이에 연속성이 존재해서는 안 됩니다(날씨, 주가 등).
- **모집단의 분포:** 모집단의 분포가 정규분포가 아니어도 되지만, 모집단이 극단적이면 표본 크기가 더 커야 할 수 있습니다.
- **유한한 평균과 분산:** 모집단의 평균과 분산이 유한해야 합니다.


## 3. 활용

#### 1)표본 평균 추정

- 중심극한정리를 사용하면 모집단의 분포가 정규분포가 아니더라도, 충분히 큰 표본 크기를 가진 경우 표본 평균은 정규분포를 따릅니다. 이를 통해 표본 평균을 기반으로 모집단 평균을 추정할 수 있습니다.

#### 2)신뢰 구간 추정(Confidence Intervals)

- 표본 평균과 표준 오차를 사용하여 모집단 평균의 신뢰 구간을 계산할 수 있습니다. 중심극한정리를 통해 표본 평균의 분포가 정규분포에 근접하므로, 신뢰 구간을 정규분포를 기반으로 계산할 수 있습니다.

  예를 들어, 표본 평균 X̄와 표준편차 σ를 알고 있을 때, 95% 신뢰 구간은 다음과 같이 계산됩니다:

  $$
  신뢰 구간 = (X̄ - Z₀.₀₂₅ * σ / √n, X̄ + Z₀.₀₂₅ * σ / √n)
  $$

  여기서 Z₀.₀₂₅는 95% 신뢰수준에 해당하는 Z-값입니다.

#### 3)가설 검정 (Hypothesis Testing)

- 중심극한정리를 활용하여, 표본 평균을 사용해 모집단 평균에 대한 가설을 검정할 수 있습니다. 예를 들어, t-검정과 z-검정 같은 방법을 사용하여 두 집단 간의 차이나 평균의 차이를 검정할 수 있습니다.

  예를 들어, 다음과 같은 가설을 검정할 수 있습니다:

  - **귀무가설**: 모집단 평균 μ₀와 표본 평균 X̄를 비교하는 경우
  - **대립가설**: 두 집단의 평균이 유의미하게 다른지 검정

#### 4)표본 크기 결정

- 연구에서 원하는 정확도와 신뢰도를 달성하기 위해 필요한 표본 크기를 결정할 때 중심극한정리를 활용할 수 있습니다. 표본 크기가 클수록 표본 평균의 분포는 모집단 평균에 더 가까워집니다.


## 4. 실무 및 데이터 분석에서의 활용

#### 1)데이터 분석 및 모델링

- **데이터 분석**: 데이터가 정규분포를 따르지 않더라도, 표본의 크기가 충분히 크면 표본 평균의 분포는 정규분포에 가까워집니다. 중심극한정리를 활용하면, 비정규적인 데이터라도 정규분포 기반의 분석 기법을 적용할 수 있습니다. 이는 데이터 분석에서 유연성을 높이고, 다양한 통계적 방법을 적용할 수 있게 합니다.

  예를 들어, 데이터가 비정규 분포를 가진 경우에도, 표본 크기가 커지면 평균의 분포는 정규분포에 수렴합니다. 따라서, 표본 평균을 사용하여 신뢰 구간을 추정하거나 가설 검정을 수행할 수 있습니다. 이는 실험적 데이터나 설문조사 데이터 등에서 유용하게 활용됩니다.

- **모델링**: 많은 통계적 모델과 기법은 데이터가 정규분포를 따른다고 가정합니다. 중심극한정리는 표본 평균이 정규분포에 가까워지므로 이러한 모델을 적용하는 데 도움을 줍니다. 예를 들어, 회귀 분석이나 ANOVA와 같은 기법은 데이터가 정규분포를 따를 때 효과적이지만, 표본의 크기가 충분히 크면 정규성의 가정을 완화할 수 있습니다.

#### 2)품질 관리

- **품질 모니터링**: 제조 공정이나 서비스의 품질을 모니터링하고 개선하기 위해 표본 데이터를 수집하여 분석합니다. 중심극한정리는 품질 분석에서 중요한 역할을 합니다. 예를 들어, 생산된 제품의 품질 특성을 측정하여 평균과 분산을 추정하고, 이를 기준으로 공정이 설정된 품질 기준에 부합하는지 평가할 수 있습니다.

  - **품질 특성의 추정**: 공정의 품질 특성(예: 평균 길이, 평균 무게 등)을 추정할 때, 표본 평균을 계산합니다. 중심극한정리에 따라, 충분히 큰 표본을 사용하면 표본 평균의 분포는 정규분포에 가까워지므로, 이를 기반으로 품질 특성을 추정할 수 있습니다.

  - **공정 개선**: 품질 관리를 위해 표본 데이터를 분석하여 공정의 평균과 분산을 추정하고, 공정이 품질 기준을 충족하는지 평가합니다. 공정이 기준에 미치지 못하는 경우, 이를 조정하여 품질을 개선할 수 있습니다.

  - **공정 능력 분석**: 공정 능력 지표(예: Cp, Cpk 등)를 계산할 때도 중심극한정리를 활용할 수 있습니다. 이러한 지표들은 공정의 품질을 평가하는 데 사용되며, 표본 데이터의 평균과 분산을 기반으로 계산됩니다. 중심극한정리에 의해 표본 평균의 분포가 정규분포에 가까워지므로, 이러한 지표의 계산과 해석이 정확해집니다.

**요약**:
- **데이터 분석**: 비정규 데이터라도 중심극한정리를 통해 정규분포 기반의 분석 기법을 적용할 수 있으며, 모델링에서도 유용하게 활용됩니다.
- **품질 관리**: 품질 특성을 추정하고 공정 개선을 위해 표본 데이터를 분석할 때 중심극한정리를 활용하여 공정의 평균과 분산을 평가하고, 공정이 기준에 부합하는지 판단할 수 있습니다.

## 코드 실습

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

# Parameters
population_size = 10000  # 전체 데이터의 크기
sample_size = 30         # 각 표본의 크기
num_samples = 1000       # 추출할 표본의 개수

# 비정규 분포 데이터 생성 (예: 균등 분포)
population = np.random.uniform(low=0, high=10, size=population_size)

# 표본 평균 계산
sample_means = [np.mean(np.random.choice(population, size=sample_size, replace=False)) for _ in range(num_samples)]

# 결과 시각화
plt.figure(figsize=(14, 6))

# 원본 데이터 분포 시각화
plt.subplot(1, 3, 1)
plt.hist(population, bins=30, edgecolor='black', alpha=0.7)
plt.title('Original Population Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')

# 표본 평균의 분포와 밀도 함수 시각화
plt.subplot(1, 3, 2)
sns.histplot(sample_means, bins=30, kde=True, color='skyblue', stat='density')
plt.title('Distribution of Sample Means')
plt.xlabel('Mean Value')
plt.ylabel('Density')

# 정규 Q-Q 플롯
plt.subplot(1, 3, 3)
stats.probplot(sample_means, dist="norm", plot=plt)
plt.title('Q-Q Plot of Sample Means')

plt.tight_layout()
plt.show()
```
