# 시계열 데이터의 통계적 분석

## 1. 개요
- 시계열 데이터는 시간에 따라 수집된 데이터를 말하며, 이러한 데이터는 특정 시간 간격으로 측정된 값들의 집합입니다.
- 시계열 데이터의 분석은 데이터의 추세, 계절성, 불규칙성을 파악하고, 이를 바탕으로 미래를 예측하는 데 중요한 역할을 합니다.
- 주식 시장, 기상 예측, 경제 지표 분석 등 다양한 분야에서 활용됩니다.

## 2. 시계열 데이터의 특징

### 1) 자기상관성과 자기상관계수
- 자기상관성은 시계열 데이터에서 관측치들이 시간의 순서에 따라 서로 상관관계를 가지는 현상을 의미합니다.
- 자기상관계수(ACF, Autocorrelation Function)는 특정 시점과 일정 시차(lag)를 가진 시점 사이의 상관관계를 측정한 값으로, 시계열 데이터의 패턴을 이해하는 데 중요한 지표입니다. 자기상관성이 높다면, 과거 데이터가 미래 데이터 예측에 유의미한 정보를 제공할 수 있음을 의미합니다.

#### ACF (Autocorrelation Function)
- ACF는 시계열 데이터의 자기상관성을 시차(lag)에 따라 측정하여, 시간에 따른 데이터를 분석하는 데 사용됩니다. ACF는 다음과 같이 정의됩니다:

$$
\rho(k) = \frac{\sum_{t=1}^{N-k} (y_t - y)(y_{t+k} - y)}{\sum_{t=1}^{N} (y_t - y)^2}
$$

여기서, k는 시차(lag), yt는 t시점의 데이터, y는 시계열 데이터의 평균입니다.

#### PACF (Partial Autocorrelation Function)
- PACF는 특정 시차에서 직전 시점과의 직접적인 상관관계를 나타내는 함수입니다. ACF와 달리, PACF는 중간 시점들의 영향을 제거하여 순수한 상관성을 측정합니다. PACF는 다음과 같이 정의됩니다:

$$
\phi_{kk} = \text{corr}(y_t, y_{t-k} \mid y_{t-1}, \dots, y_{t-k+1})
$$

PACF는 특정 시차에서 자기상관이 더 이상 유의미하지 않다고 판단할 수 있는 기준을 제공합니다.

### 2) 추세
- 추세는 시간에 따라 데이터가 지속적으로 증가하거나 감소하는 경향을 말합니다. 예를 들어, 인구 증가, 물가 상승 등이 있을 수 있습니다.
- 추세는 장기적인 패턴을 나타내며, 이를 파악하면 전체적인 데이터 흐름을 이해할 수 있습니다.

### 3) 계절성
- 계절성은 일정한 주기마다 반복되는 패턴을 의미합니다. 예를 들어, 계절에 따른 매출 변동, 시간대별 교통량 변화 등이 있습니다.
- 계절성은 예측 모형에서 중요한 요소로, 이를 정확히 반영하는 것이 필요합니다.

### 4) 불규칙성
- 불규칙성은 추세나 계절성으로 설명되지 않는 비정상적이고 무작위적인 변동을 말합니다.
- 이러한 변동은 예측하기 어려우며, 잔차 분석 등을 통해 이러한 요소를 파악할 수 있습니다.

## 3. 시계열 분석의 자료

### 1) 비정상 시계열 데이터와 데이터 변환
- 비정상 시계열 데이터는 평균, 분산, 공분산이 시간이 지남에 따라 변화하는 데이터로, 안정적이지 않은 데이터를 말합니다.
- 비정상 데이터를 분석하기 위해서는 데이터를 변환하여 정상성을 확보하는 과정이 필요합니다.

  - **차분**: 차분은 시계열 데이터의 값을 이전 시점의 값과 비교하여 차이를 계산하는 방법으로, 주로 추세를 제거하거나 안정성을 확보하기 위해 사용됩니다.

    $$ \Delta y_t = y_t - y_{t-1} $$

  - **로그변환**: 로그변환은 데이터의 분산을 안정화시키고, 비정상성을 완화하기 위해 사용됩니다. 주로 데이터 값의 크기가 큰 경우 유용합니다.

    $$ y_t' = \log(y_t) $$

### 2) 정상 시계열 데이터의 특징
정상 시계열 데이터는 시간에 따라 일정한 통계적 성질을 유지하는 데이터를 말합니다. 정상성을 확보한 데이터는 예측과 분석에 유리합니다.

- **일정한 평균**: 시계열 데이터의 평균이 시간에 따라 변하지 않는 것을 말합니다. 즉, 장기적으로 데이터가 일정한 평균을 중심으로 변동합니다.
- **일정한 분산**: 데이터의 변동 폭이 시간에 따라 일정한 상태를 유지하는 것을 의미합니다. 즉, 분산이 일정해야 분석 결과의 신뢰성을 높일 수 있습니다.
- **일정한 공분산**: 시차(lag)가 같은 두 시점 간의 공분산이 시간에 따라 일정한 것을 의미합니다. 이는 자기상관성을 분석할 때 중요한 요소입니다.

## 4. 시계열분석기법

### 1) 자료 형태에 따른 분류
시계열 분석은 자료의 형태에 따라 다양한 방법으로 분류될 수 있습니다. 주요하게는 계절성, 추세, 불규칙성을 반영한 모델과, 이러한 요소를 반영하지 않는 모델로 나눌 수 있습니다.

#### 일변량 시계열 분석
- **정의**: 일변량 시계열 분석은 하나의 시계열 데이터 변수만을 사용하는 분석 방법입니다. 이 방법은 단일 변수의 시간에 따른 패턴을 분석하고 예측하는 데 초점을 맞춥니다.
- **특징**: 데이터는 시간에 따라 순서가 있으며, 자기상관성, 추세, 계절성 등을 분석하여 미래의 값을 예측합니다.
- **예시**:
  - **주식 가격 예측**: 특정 주식의 가격 변동을 분석하여 향후 가격을 예측합니다.
  - **기온 변화 예측**: 일일 기온 데이터를 사용하여 미래 기온을 예측합니다.
  - **판매량 분석**: 특정 제품의 월별 판매량을 분석하여 향후 판매량을 예측합니다.

#### 다중 시계열 분석
- **정의**: 다중 시계열 분석은 두 개 이상의 시계열 데이터를 동시에 분석하여 변수 간의 관계를 이해하고 예측하는 방법입니다. 이 방법은 변수 간의 상호작용과 공동 변동성을 분석하는 데 중점을 둡니다.
- **특징**: 데이터는 시간에 따라 여러 변수의 변동을 포함하고 있으며, 변수 간의 관계를 모델링하여 더 정확한 예측을 시도합니다.
- **예시**:
  - **경제 지표 분석**: GDP, 실업률, 물가상승률 등의 여러 경제 지표를 동시에 분석하여 경제 상태를 예측합니다.
  - **기상 데이터 분석**: 기온, 습도, 강수량 등 여러 기상 요소를 분석하여 날씨 예측을 수행합니다.
  - **마케팅 캠페인 효과 분석**: 광고비, 판매량, 소비자 행동 등 여러 변수의 시계열 데이터를 분석하여 캠페인의 효과를 평가합니다.

### 2) 이동평균법
일정한 기간별로 이동 평균을 계산하고 이를 기반으로 추세를 파악해서 예측하는 방법입니다.
- 시간이 지남에 따라 사용하는 데이터가 달라지지만, 평균을 구하기 때문에 계절성에 따른 변동성 및 불규칙 변동이 적습니다.
- 변동성을 제거하기에 예측을 쉽게 할 수 있지만, 최근 자료에 대한 가중치가 없다는 단점이 있습니다.
- 단순 이동평균, 가중 이동평균 등 다양한 변형이 있습니다.

### 3) 지수평활법
- 모든 시계열 자료를 사용해 평균을 구하는 방법입니다. 
- 최근 데이터에 더 높은 가중치를 부여하여 최근의 변동을 더 민감하게 반영합니다.
- 자료 수가 많고 안정적인 패턴을 보일수록 예측력이 상승하고 불규칙 변동의 영향력이 줄어듭니다. 주로 중기 이상의 데이터에 대해 사용합니다.
- Holt-Winters 방법은 추세와 계절성을 함께 고려한 지수평활법의 대표적인 예입니다.

## 5. 시계열분석 모델

### 1) AR(자기회귀모형)
- AR 모델은 현재 시점의 데이터가 이전 시점 데이터의 가중합으로 설명될 수 있는 모델입니다. 자기상관성을 이용하여 미래 데이터를 예측합니다.
- t시점의 값 yt는 이전 시점들 yt-1, yt-2, ... , yt-p 에 의해 설명 가능합니다:

  $$
  y_t = \phi_1 y_{t-1} + \phi_2 y_{t-2} + \dots + \phi_p y_{t-p} + \epsilon_t
  $$

  여기서 phi_i는 자기회귀 계수이고, epsilon_t는 백색 잡음입니다.

- ACF는 빠르게 감소하고, PACF는 특정 시점에서 절단점을 보입니다.

### 2) MA(이동평균모형)
- MA 모델은 현재 시점의 데이터가 이전 시점의 오차들의 가중합으로 설명되는 모델입니다:

  $$
  y_t = \epsilon_t + \theta_1 \epsilon_{t-1} + \theta_2 \epsilon_{t-2} + \dots + \theta_q \epsilon_{t-q}
  $$

  여기서 theta_i는 이동평균 계수이고, epsilon_t는 백색 잡음입니다.

- ACF가 특정 시점에서 절단점을 보이고, PACF는 빠르게 감소합니다.

### 3) ARIMA(자기회귀누적평균이동모형)
- AR과 MA의 결합한 형태이자 비정상 시계열 모델로, 차분 혹은 변환으로 AR, MA, ARMA로 정상화합니다.
- ARIMA 모델은 다음과 같이 표현됩니다:

  $$
  y_t' = \phi_1 y_{t-1}' + \dots + \phi_p y_{t-p}' + \epsilon_t + \theta_1 \epsilon_{t-1} + \dots + \theta_q \epsilon_{t-q}
  $$

  여기서 y_t'는 차분된 데이터입니다.

- 정상화:
  - p : AR의 차수
  - q : MA의 차수
  - d : ARIMA에서 ARMA로 정상화할 때의 차분 횟수

### 4) SARIMA (Seasonal ARIMA)
- SARIMA는 ARIMA 모델에 계절성을 추가한 모델입니다. 계절성을 가진 시계열 데이터를 분석하는 데 효과적입니다. SARIMA는 계절적 차분과 계절적 AR, MA 항을 추가하여 계절성 패턴을 더욱 정확하게 반영합니다.

### 5) Exponential Smoothing Models
- 이 모델은 Holt-Winters 등 다양한 형태가 있으며, 추세와 계절성을 함께 고려하여 시계열 데이터를 예측하는 방법입니다. 지수평활법을 기반으로 하여 데이터의 가중치를 점진적으로 감소시킴으로써 더 최근의 데이터에 더 큰 가중치를 부여합니다.

### 6) GARCH (Generalized Autoregressive Conditional Heteroskedasticity)
- GARCH 모델은 시계열 데이터의 변동성이 시간에 따라 변화하는 경우, 특히 금융 데이터의 변동성을 분석하는 데 사용됩니다. 시계열 데이터의 이분산성을 고려한 모델로, 데이터의 변동성이 클 때 더욱 유용하게 적용됩니다.

### 7) Prophet
- Prophet은 Facebook에서 개발한 시계열 예측 도구로, 복잡한 패턴과 결측치를 처리하는 데 강점을 가지고 있습니다. 추세와 계절성, 휴일 효과 등을 반영하여 유연하게 예측을 수행할 수 있으며, 사용자에게 친화적인 인터페이스를 제공합니다.

## 6. 모델평가 및 진단

- **잔차 분석**: 잔차는 실제 값과 예측 값의 차이를 나타내며, 잔차 분석은 모델이 적합하게 데이터 패턴을 반영했는지 확인하는 과정입니다. 잔차가 백색잡음(white noise)처럼 무작위적으로 분포해야 모델이 적합하다고 볼 수 있습니다.

- **AIC/BIC**: Akaike 정보 기준(AIC)과 Bayesian 정보 기준(BIC)은 모델의 복잡도와 적합도를 평가하는 지표입니다. 값이 낮을수록 더 적합한 모델로 간주되며, AIC와 BIC를 통해 여러 모델 간 비교가 가능합니다.

- **예측 성능 평가**: 예측 성능을 평가하기 위해 RMSE(Root Mean Square Error)와 MAE(Mean Absolute Error) 같은 지표를 사용합니다. 이들 지표는 예측 값과 실제 값 간의 차이를 나타내며, 값이 낮을수록 모델의 예측 정확도가 높음을 의미합니다.

```python
# 필요한 라이브러리 임포트
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima_model import ARIMA
from statsmodels.tsa.stattools import adfuller

# 데이터 불러오기 (여기서는 예제 데이터를 생성합니다)
# 실제 데이터가 있다면, pd.read_csv() 등을 사용하여 불러옵니다.
np.random.seed(0)
data = np.random.randn(100).cumsum()  # 누적합으로 시계열 데이터 생성
dates = pd.date_range(start='2020-01-01', periods=100, freq='D')
series = pd.Series(data, index=dates)

# 데이터 시각화
plt.figure(figsize=(10, 6))
plt.plot(series, label='Original Data')
plt.title('Time Series Data')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.show()

# ARIMA 모델을 적용하기 전에 데이터가 정상성(Stationarity)을 가지는지 확인합니다.
result = adfuller(series)
print('ADF Statistic:', result[0])
print('p-value:', result[1])

# 데이터가 비정상적이라면 차분을 통해 정상성을 확보합니다.
# 여기서는 비정상 데이터가 아니라고 가정합니다.

# ARIMA 모델을 적합합니다 (p, d, q의 값은 예시로 설정한 것)
model = ARIMA(series, order=(5, 1, 0))  # p=5, d=1, q=0
model_fit = model.fit(disp=0)

# 모델 요약
print(model_fit.summary())

# 예측 수행
forecast = model_fit.forecast(steps=10)[0]

# 예측 결과 시각화
plt.figure(figsize=(10, 6))
plt.plot(series, label='Original Data')
plt.plot(pd.date_range(start=series.index[-1] + pd.Timedelta(days=1), periods=10, freq='D'), forecast, label='Forecast', color='red')
plt.title('ARIMA Forecast')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.show()
```
