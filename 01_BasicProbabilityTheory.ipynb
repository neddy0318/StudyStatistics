# 확률 이론 기초 Basic Probability Theory

## 1. 기본 정의

&nbsp;&nbsp; - 특정한 사건이 일어날 가능성을 수치로 표현한 것으로, 특정 사건 A가 일어날 확률을 통상 P(A)로 표현합니다.</br>
&nbsp;&nbsp; - `0 ≤ P(A) ≤ 1`

## 2. 표본공간과 확률공간

#### 1)표본공간 (Sample Space)

&nbsp;&nbsp; - 표본공간은 실험이나 사건의 모든 가능한 결과들의 집합을 의미합니다. 예를 들어, 동전을 던지는 실험에서 표본공간은 {앞면, 뒷면}입니다.</br>

&nbsp;&nbsp; - 보통 S 또는 Ω로 표기합니다.

&nbsp;&nbsp; **예시:**</br>
&nbsp;&nbsp;&nbsp; - 동전 던지기: Ω = { 앞면, 뒷면 }</br>

&nbsp;&nbsp;&nbsp; - 주사위 던지기: Ω = { 1, 2, 3, 4, 5, 6 }</br>

&nbsp;&nbsp;&nbsp; - 두 개의 주사위를 던지기: Ω = { (1,1), (1,2), …, (6,5), (6,6) }


#### 2)확률공간 (Probability Space)

&nbsp;&nbsp; - 확률공간은 표본공간을 기반으로 확률을 정의하는 보다 구조화된 개념입니다. 확률공간은 세 가지 요소로 구성됩니다:</br>

&nbsp;&nbsp;&nbsp; **(1)표본공간 Ω**: 모든 가능한 결과들의 집합.</br>

&nbsp;&nbsp;&nbsp; **(2)사건의 시그마 대수 F**: 표본공간의 부분집합들로 이루어진 시그마 대수. 이 부분집합들은 사건(event)이라고 불립니다. 사건의 집합이 폐쇄적인 집합 연산(예: 여집합, 합집합)에 대해 닫혀 있는 성질을 가져야 합니다.</br>

&nbsp;&nbsp;&nbsp; **(3)확률 측도 P**: 시그마 대수의 각 사건에 확률을 부여하는 함수. P: F → [0, 1]로 정의됩니다. P(Ω) = 1이고, 시그마 대수의 각 사건에 대해 0에서 1 사이의 값을 가집니다.


&nbsp;&nbsp; #예시:</br>

&nbsp;&nbsp;&nbsp; (1)동전 던지기:</br>

&nbsp;&nbsp;&nbsp;&nbsp; - 표본공간: Ω = { 앞면, 뒷면 }</br>

&nbsp;&nbsp;&nbsp;&nbsp; - 시그마 대수: F = { ∅, {앞면}, {뒷면}, Ω }</br>

&nbsp;&nbsp;&nbsp;&nbsp; - 확률 측도: P({앞면}) = 0.5, P({뒷면}) = 0.5


&nbsp;&nbsp;&nbsp; (2)주사위 던지기:</br>

&nbsp;&nbsp;&nbsp;&nbsp; - 표본공간: Ω = { 1, 2, 3, 4, 5, 6 }</br>

&nbsp;&nbsp;&nbsp;&nbsp; - 시그마 대수: 주사위 던지기의 경우 모든 부분집합이 시그마 대수의 원소가 될 수 있습니다.</br>

&nbsp;&nbsp;&nbsp;&nbsp;- 확률 측도: P({i}) = 1/6 (i = 1, 2, 3, 4, 5, 6)



## 3. 기본적인 확률 계산

#### 1)조건부확률과 결합확률

&nbsp;&nbsp; (1)조건부확률 (Conditional Probability)</br>

&nbsp;&nbsp;&nbsp; - 조건부확률은 어떤 사건 B가 발생한 조건 하에서 다른 사건 A가 발생할 확률을 의미합니다. 이는 사건 B가 발생했을 때 사건 A가 발생할 확률을 나타냅니다.</br>

&nbsp;&nbsp;&nbsp; - P(A|B)로 표기합니다.</br>

&nbsp;&nbsp;&nbsp; - 정의 : `P(A|B) = P(A ∩ B) / P(B), 단 P(B) > 0`</br>

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; - P(A|B)는 B가 발생했을 때 A가 발생할 조건부확률입니다.</br>

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; - P(A ∩ B)는 A와 B가 모두 발생할 확률입니다.</br>

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; - P(B)는 B가 발생할 확률입니다.</br>


&nbsp;&nbsp; **예시:** 주사위를 던질 때, 짝수가 나왔다는 조건 하에서 4가 나올 확률:</br>

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; - A: 4가 나오는 사건</br>

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; - B: 짝수가 나오는 사건 (2, 4, 6)</br>

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; - P(A) = 1/6, P(B) = 1/2</br>

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; - P(A ∩ B) = 1/6</br>

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; - P(A|B) = (1/6) / (1/2) = 1/3</br>


&nbsp;&nbsp; (2)결합 확률 (Joint Probability)</br>

&nbsp;&nbsp;&nbsp; - 결합 확률은 두 개 이상의 사건이 동시에 발생할 확률을 의미합니다. 이는 사건 A와 사건 B가 동시에 발생하는 경우를 고려합니다.</br>

&nbsp;&nbsp;&nbsp; - P(A ∩ B)로 표기합니다.</br>

&nbsp;&nbsp;&nbsp; - 정의 : `P(A ∩ B)`

&nbsp;&nbsp; **예시:** 주사위를 두 번 던질 때, 첫 번째 주사위가 4가 나오고, 두 번째 주사위가 6이 나올 확률.</br>

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; - A: 첫 번째 주사위가 4가 나오는 사건</br>

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; - B: 두 번째 주사위가 6이 나오는 사건</br>

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; - P(A ∩ B) = P(A) * P(B) = 1/6 * 1/6 = 1/36


&nbsp;&nbsp; (3) 조건부 확률과 결합 확률의 관계</br>

&nbsp;&nbsp; -조건부 확률과 결합 확률은 다음과 같은 관계가 있습니다:</br>

&nbsp;&nbsp;&nbsp; P(A ∩ B) = P(A|B) * P(B)</br>

&nbsp;&nbsp;&nbsp; P(A ∩ B) = P(B|A) * P(A)

&nbsp;&nbsp; **예시:** 주사위를 두 번 던질 때, 첫 번째 주사위가 4가 나왔을 때 두 번째 주사위가 6이 나올 확률을 조건부 확률과 결합 확률을 통해 계산할 수 있습니다.</br>

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; - 조건부 확률: P(A|B) = 1/6 (첫 번째 주사위가 4가 나왔을 때 두 번째 주사위가 6이 나올 확률)</br>

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; - 결합 확률: P(A ∩ B) = P(A|B) * P(B) = 1/6 * 1/6 = 1/36



#### 2)합의법칩과 곱의 법칙</br>

&nbsp;&nbsp; (1)합의법칙 : 두 사건 A와 B가 서로 배타적이라면(동시에 일어나지 않는다면), A 또는 B가 일어날 확률은 `P(A) + P(B)` 입니다.</br>

&nbsp;&nbsp; (2)곱의법칙 : 두 사건 A와 B가 서로 독립적이라면(서로에게 영향을 주지 않는다면), A와 B가 동시에 발생할 확률은 `P(A) * P(B)` 입니다.


## 4. 통계적 파라미터

&nbsp;&nbsp; - 통계적 파라미터는 모집단의 특성을 나타내는 수치적 값으로, 모집단의 분포를 설명하거나 요약하는 데 사용됩니다. 주요 통계적 파라미터는 평균, 중앙값, 분산, 표준편차, 모멘트 등이 있습니다.

&nbsp;&nbsp; 1)평균(Mean) : 평균은 모집단의 값들을 모두 합한 후 값의 개수로 나눈 값입니다. 이는 모집단의 중심 경향을 나타내는 대표적인 값입니다.</br>

&nbsp;&nbsp;&nbsp;&nbsp; - 모집단 평균: `μ = (ΣX) / N`</br>

&nbsp;&nbsp;&nbsp;&nbsp; - 표본 평균: `x̄ = (Σx) / n`</br>

&nbsp;&nbsp;&nbsp;&nbsp; (N은 모집단의 크기, n은 표본의 크기, X는 모집단의 값, x는 표본의 값을 의미합니다)


&nbsp;&nbsp; 2)중앙값(Median) : 중앙값은 데이터를 크기순으로 정렬했을 때, 중앙에 위치한 값입니다. 데이터가 짝수 개일 경우, 중앙에 위치한 두 값의 평균을 중앙값으로 합니다. 극단값에 영향을 받지 않습니다.


&nbsp;&nbsp; 3)분산(Variance) : 분산은 데이터 값들이 평균에서 얼마나 떨어져 있는지를 나타내는 값입니다. 분산이 크면 데이터 값들이 평균에서 멀리 떨어져 있음을 의미합니다.</br>

&nbsp;&nbsp;&nbsp;&nbsp; - 모집단 분산: `σ² = Σ(X - μ)² / N`</br>

&nbsp;&nbsp;&nbsp;&nbsp; - 표본 분산: `s² = Σ(x - x̄)² / (n - 1)`


&nbsp;&nbsp; 4)표준편차(Standard Deviation) : 표준편차는 분산의 제곱근으로, 데이터 값들이 평균에서 얼마나 떨어져 있는지를 나타내는 값입니다. 분산과 마찬가지로 데이터의 흩어짐 정도를 측정합니다.</br>

&nbsp;&nbsp;&nbsp;&nbsp; - 모집단 표준편차: `σ = sqrt(σ²)`</br>

&nbsp;&nbsp;&nbsp;&nbsp; - 표본 표준편차: `s = sqrt(s²)`


&nbsp;&nbsp; 5)모멘트(Moments) : 모멘트는 데이터의 분포 특성을 나타내는 값으로, 평균, 분산, 비대칭도, 첨도를 포함합니다. n차 모멘트는 데이터를 n제곱하여 평균을 계산한 값입니다.</br>

&nbsp;&nbsp;&nbsp;&nbsp; - n차 모멘트: `μ'_n = E[(X - μ)ⁿ]` (여기서 E는 기대값을 의미)


&nbsp;&nbsp; 6)기대값 (Expectation) : 기대값은 확률 변수의 평균 값을 의미합니다. 이는 확률 분포의 중심 경향을 나타내는 값입니다.</br>

&nbsp;&nbsp;&nbsp;&nbsp; - 이산형 확률 변수: `E(X) = Σ [x * P(x)]`</br>

&nbsp;&nbsp;&nbsp;&nbsp; - 연속형 확률 변수: `E(X) = ∫ x * f(x) dx`</br>

&nbsp;&nbsp;&nbsp;&nbsp; (여기서 P(x)는 확률 질량 함수, f(x)는 확률 밀도 함수)


&nbsp;&nbsp; 7)비대칭도(Skewness) : 비대칭도는 데이터의 비대칭 정도를 나타내는 값입니다. 비대칭도가 0이면 데이터가 대칭 분포를 따르며, 양수면 오른쪽으로 긴 꼬리를, 음수면 왼쪽으로 긴 꼬리를 가집니다.

&nbsp;&nbsp; 8)첨도(Kurtosis) : 첨도는 데이터의 꼬리 두께를 나타내는 값입니다. 첨도가 높으면 데이터의 꼬리가 두껍고, 낮으면 꼬리가 얇습니다.



## 5. 공분산과 상관관계

#### 1)공분산(Covariance)</br>

&nbsp;&nbsp; -공분산은 두 확률 변수 간의 관계를 나타내는 값으로, 한 변수의 변화에 따라 다른 변수가 어떻게 변하는지를 측정합니다. 양의 공분산은 두 변수가 같은 방향으로 움직이는 경향이 있음을, 음의 공분산은 반대 방향으로 움직이는 경향이 있음을 의미합니다.</br>

&nbsp;&nbsp;&nbsp; - 기호 : 공분산: Cov(X, Y)</br>

&nbsp;&nbsp;&nbsp; - 모집단 공분산: Cov(X, Y) = E[(X - μ_X)(Y - μ_Y)]</br>

&nbsp;&nbsp;&nbsp; - 표본 공분산: Cov(X, Y) = Σ[(x_i - x̄)(y_i - ȳ)] / (n - 1)</br>

&nbsp;&nbsp;&nbsp; (여기서 E는 기대값, μ_X와 μ_Y는 각각 X와 Y의 평균, x̄와 ȳ는 각각 표본 X와 Y의 평균, n은 표본의 크기를 의미)


&nbsp;&nbsp; **특징:**</br>

&nbsp;&nbsp;&nbsp; - Cov(X, Y) > 0: X와 Y가 같은 방향으로 움직임</br>

&nbsp;&nbsp;&nbsp;- Cov(X, Y) < 0: X와 Y가 반대 방향으로 움직임</br>

&nbsp;&nbsp;&nbsp;- Cov(X, Y) = 0: X와 Y 사이에 선형적인 관계가 없음


&nbsp;&nbsp; **예시:** 두 변수 X와 Y의 값이 다음과 같을 때 공분산을 계산합니다.</br>

&nbsp;&nbsp;&nbsp; - X: [1, 2, 3, 4, 5]</br>

&nbsp;&nbsp;&nbsp; - Y: [2, 4, 6, 8, 10]</br>

&nbsp;&nbsp;&nbsp; - 표본 공분산: Cov(X, Y) = Σ[(x_i - x̄)(y_i - ȳ)] / (n - 1) = 5


#### 2)상관관계(Correlation)</br>

&nbsp;&nbsp;&nbsp; - 상관관계는 두 변수 간의 선형 관계의 강도와 방향을 나타내는 값으로, 공분산을 각 변수의 표준편차로 나누어 계산합니다. 상관관계는 -1에서 1 사이의 값을 가지며, 1에 가까울수록 강한 양의 상관관계를, -1에 가까울수록 강한 음의 상관관계를 나타냅니다. 0에 가까운 값은 선형적인 관계가 없음을 의미합니다.</br>

&nbsp;&nbsp;&nbsp; - 상관계수: ρ(X, Y) 또는 r(X, Y)</br>

&nbsp;&nbsp;&nbsp; - 모집단 상관계수: ρ(X, Y) = Cov(X, Y) / (σ_X * σ_Y)</br>

&nbsp;&nbsp;&nbsp; - 표본 상관계수: r(X, Y) = Cov(X, Y) / (s_X * s_Y)</br>

&nbsp;&nbsp;&nbsp; (여기서 σ_X와 σ_Y는 각각 X와 Y의 표준편차, s_X와 s_Y는 각각 표본 X와 Y의 표준편차를 의미)


&nbsp;&nbsp; **특징:**</br>

&nbsp;&nbsp;&nbsp; - ρ(X, Y) = 1: X와 Y가 완벽한 양의 선형 관계를 가짐</br>

&nbsp;&nbsp;&nbsp; - ρ(X, Y) = -1: X와 Y가 완벽한 음의 선형 관계를 가짐</br>

&nbsp;&nbsp;&nbsp; - ρ(X, Y) = 0: X와 Y 사이에 선형적인 관계가 없음


&nbsp;&nbsp; **예시:** 두 변수 X와 Y의 값이 다음과 같을 때 상관계수를 계산합니다.</br>

&nbsp;&nbsp;&nbsp; - X: [1, 2, 3, 4, 5]</br>

&nbsp;&nbsp;&nbsp; - Y: [2, 4, 6, 8, 10]</br>

&nbsp;&nbsp;&nbsp; - 표본 상관계수: r(X, Y) = Cov(X, Y) / (s_X * s_Y) = 1


#### 3)공분산과 상관관계의 차이</br>

&nbsp;&nbsp;&nbsp; - **단위:** 공분산은 두 변수의 단위를 곱한 단위를 가지므로, 단위에 따라 값이 달라질 수 있습니다. 반면, 상관관계는 단위가 없는 값으로, -1에서 1 사이의 범위를 가지며 단위에 무관하게 비교 가능합니다.</br>

&nbsp;&nbsp;&nbsp; - **해석:** 공분산은 두 변수의 방향성을 나타내지만, 그 크기는 해석하기 어렵습니다. 상관관계는 두 변수 간의 관계의 강도와 방향을 명확히 해석할 수 있습니다.



## 7. 확률분포

&nbsp;&nbsp; - 확률변수가 취할 수 있는 값들과 그 값들이 발생하는 확률을 설명하는 개념으로, 크게 이산확률분포와 연속확률분포로 나눌 수 잇습니다.

#### 1)이산확률분포</br>

&nbsp;&nbsp; - 확률 변수가 취할 수 잇는 값이 개별적이고 셀 수 있는 경우를 말합니다.</br>

&nbsp;&nbsp; - 주사위 던지기나 동전 던지기의 확률 변수를 예로 들 수 있습니다.</br>

&nbsp;&nbsp; - 베르누이 분포, 이항 분포, 포아송 분포 등이 있습니다.


#### 2)연속확률분포</br>

&nbsp;&nbsp; - 확률 변수가 연속적인 범위의 값(실수 범위의 값)을 취할 수 있는 경우를 말합니다.</br>

&nbsp;&nbsp; - 연속 확률분포는 연속적인 확률 변수가 취할 수 있는 값에 대해 확률 밀도 함수를 사용하여 확률을 할당합니다. 연속 확률 변수는 특정 구간 내의 모든 값을 취할 수 있습니다.</br>
&nbsp;&nbsp; - 확률이기 때문에 모든 확률 변수 전 구간의 적분 값은 1이다.</br>

&nbsp;&nbsp; - 전 국민의 키나 몸무게에 대한 분포 등을 예로 들 수 있습니다.</br>

&nbsp;&nbsp; - 정규 분포, 지수 분포, 균등 분포, 감마 분포 등이 있습니다.
