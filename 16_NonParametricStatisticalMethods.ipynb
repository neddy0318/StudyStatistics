# 1. 개요

#### 1) 개요
- **모수적 방법**: 모집단의 분포(예: 정규 분포)를 가정하고 분석합니다. 이를 모수적 방법이라고 합니다.
- **비모수적 방법**: 모집단의 분포에 대해 특정한 가정을 하지 않고 분석합니다. 데이터의 분포에 대한 가정 없이 접근하는 통계적 방법입니다.

## 2) 사용해야 하는 상황 및 주의점
- 보통 데이터가 정규성 가정을 충족하지 못하면 비모수적 방법을 사용한다고 이야기하지만, 정규성을 만족시키지 못하는 경우에도 모수적 방법이 더 나을 때가 있습니다.
    - **표본수가 충분한 경우**:
        - Minitab 통계학자들의 연구 결과에 따르면, 다음과 같은 경우에는 모수적 방법이 유용할 수 있습니다:
          - One sample t-test: 20개
          - Two sample t-test: 15개
          - One-way ANOVA: 그룹별로 15~20개의 데이터가 필요
    - **분산 가정**: 비모수적 방법은 데이터 간의 동일한 분산(dispersion)을 가정해야 합니다. 반면, two sample t-test나 one-way ANOVA 같은 모수적 방법은 이분산 가정을 통해 다른 분산의 경우에도 처리할 수 있습니다.
    - **검정력**: 동일한 상황에서는 보통 모수적 방법이 비모수적 방법보다 검정력이 우수하여, 유의적 차이가 존재할 때 이를 더 잘 발견할 수 있습니다.

- 그럼에도 불구하고 비모수적 방법을 써야 하는 경우는 다음과 같습니다:
    - **데이터 분포의 비정규성**: 분석하려는 자료에 왜도(skeness)가 있어, 평균보다 중위수(median)로 더 대표할 수 있을 때
    - **표본 수가 적으면서 정규 분포를 따르지 않는 경우**
    - **서열 자료 및 이상치**:
        - Ordinal data, ranked data(순서 또는 서열 자료) 등이 있거나, 제거할 수 없는 outlier가 있는 경우
        - 통상의 모수적 방법은 연속적 data에서만 적용 가능하며, 이상치에 의해 쉽게 영향 받는 반면, 비모수적 방법은 서열 및 순서 데이터에도 적용 가능하고, 이상치의 영향을 덜 받습니다.

- **비모수적 방법의 가정**: 비모수적 방법이라도 무시할 수 없는 엄격한 가정들이 있습니다.
        - 데이터 분포의 중심이 평균으로 정확히 대표될 수 있고, 샘플 수가 충분히 크다면, 검정력이 더 좋은 모수적 방법의 적용을 고려하는 것이 좋습니다.
        - 데이터 분포의 중심이 중위수로 더 잘 대표된다면, 샘플 수가 충분히 크더라도 비모수적 방법을 적용하는 것이 더 낫습니다.

- **샘플 수가 매우 적을 경우**: 샘플 수가 매우 적다면 비모수적 방법을 적용하는 것은 무리가 있으므로, 가능한 데이터를 더 확보하도록 노력하는 것이 좋습니다.
    - 위에서 확인했듯이, 적정 샘플 수에 대한 가이드라인 자체도 매우 무리한 수준은 아닙니다 (15~20개 수준).
    - 게다가 샘플 수가 적은 상황에서 비효율적인 비모수적 방법을 적용하는 경우, 유의한 차이를 발견할 확률이 매우 낮다는 점을 기억해야 합니다.

## 2. 특징
- **분포 가정 없음**: 특정 분포를 가정하지 않고 데이터 분석을 수행합니다.
- **순위 기반 분석**: 데이터의 순위를 기반으로 분석을 진행합니다.
- **강건성**: 이상값에 강하고, 작은 샘플 크기에서도 효과적입니다.

## 3. 장단점
### 장점
- **유연성**: 다양한 형태의 데이터에 적용 가능합니다.
- **이상값에 강건**: 데이터에 이상값이 포함되어도 분석 결과에 영향을 미치지 않습니다.
- **소규모 샘플에도 적합**: 작은 표본 크기에서도 유용합니다.

### 단점
- **효율성**: 모수적 방법에 비해 효율성이 낮을 수 있습니다.
- **해석의 복잡성**: 결과 해석이 직관적이지 않을 수 있습니다.
- **정보 손실**: 데이터의 세부적인 정보를 잃을 수 있습니다.



## 4. 비모수적 통계 방법들

#### 1) 단일표본 부호검정 (Sign Test)
- 단일 표본의 중앙값이 특정 기준값과 다른지 검정합니다.
- 절차
    1. **기준값 설정**: 비교할 기준값을 정합니다.
    2. **차이 계산**: 각 관측치와 기준값 간의 차이를 계산합니다.
    3. **부호 부여**: 차이에 대해 부호를 부여합니다 (양수 또는 음수).
    4. **부호 카운트**: 각 부호의 빈도를 계산합니다.
    5. **검정 통계량 계산**: 부호의 빈도를 사용하여 검정 통계량을 계산합니다.
    6. **유의성 검정**: 유의성 수준에 따라 통계량을 검정합니다.

#### 2) 윌콕슨 부호 순위 검정 (Wilcoxon Signed-Rank Test)
- 두 관련 샘플 간의 중앙값 차이를 비교합니다.
- 절차
    1. **차이 계산**: 각 쌍의 차이를 계산합니다.
    2. **순위 매기기**: 차이의 절대값에 순위를 매깁니다.
    3. **부호 부여**: 순위에 원래의 부호를 부여합니다.
    4. **순위 합계 계산**: 양수와 음수의 순위 합계를 계산합니다.
    5. **검정 통계량 계산**: 순위 합계를 기반으로 검정 통계량을 계산합니다.
    6. **유의성 검정**: 유의성 수준에 따라 검정 통계량을 검정합니다.

#### 3) 윌콕슨 순위 합 테스트 (Wilcoxon Rank-Sum Test)
- 두 독립 샘플 간의 중앙값 차이를 비교합니다.
- 절차
    1. **데이터 결합**: 두 샘플을 결합하여 전체 데이터 세트를 만듭니다.
    2. **순위 매기기**: 전체 데이터에 순위를 매깁니다.
    3. **순위 합계 계산**: 각 샘플의 순위 합계를 계산합니다.
    4. **검정 통계량 계산**: 순위 합계를 기반으로 검정 통계량을 계산합니다.
    5. **유의성 검정**: 유의성 수준에 따라 검정 통계량을 검정합니다.

#### 4) 두 표본 부호 검정 (Sign Test for Two Samples)
- 두 독립 샘플 간의 중앙값 차이를 검정합니다.
- 절차
    1. **차이 계산**: 각 쌍의 차이를 계산합니다.
    2. **부호 부여**: 차이에 부호를 부여합니다.
    3. **부호 카운트**: 각 부호의 빈도를 계산합니다.
    4. **검정 통계량 계산**: 부호의 빈도를 사용하여 검정 통계량을 계산합니다.
    5. **유의성 검정**: 유의성 수준에 따라 통계량을 검정합니다.

#### 5) 크루스칼-왈리스 H 검정 (Kruskal-Wallis H Test)
- 세 개 이상의 독립 샘플 간의 분포 차이를 비교합니다.
- 절차
    1. **데이터 결합**: 모든 샘플을 결합하여 전체 데이터 세트를 만듭니다.
    2. **순위 매기기**: 전체 데이터에 순위를 매깁니다.
    3. **순위 합계 계산**: 각 샘플의 순위 합계를 계산합니다.
    4. **검정 통계량 계산**: 순위 합계를 기반으로 검정 통계량을 계산합니다.
    5. **유의성 검정**: 유의성 수준에 따라 검정 통계량을 검정합니다.

#### 6) 런 검정 (Runs Test)
- 데이터의 순서가 무작위인지 검정합니다.
- 절차
    1. **데이터 정렬**: 데이터를 정렬하여 '런'을 식별합니다.
    2. **런 계산**: 연속적으로 같은 값이 나타나는 구간을 런으로 계산합니다.
    3. **검정 통계량 계산**: 런의 수를 사용하여 검정 통계량을 계산합니다.
    4. **유의성 검정**: 유의성 수준에 따라 검정 통계량을 검정합니다.

#### 7) 스피어만의 순위 상관계수 (Spearman's Rank Correlation Coefficient)
- 두 변수 간의 순위 기반 상관관계를 측정합니다.
- 절차
    1. **순위 매기기**: 두 변수의 값에 순위를 매깁니다.
    2. **순위 차이 계산**: 두 변수 간의 순위 차이를 계산합니다.
    3. **상관계수 계산**: 순위 차이를 기반으로 상관계수를 계산합니다.
    4. **유의성 검정**: 유의성 수준에 따라 상관계수를 검정합니다.

#### 8) 맨-휘트니 U 검정 (Mann-Whitney U Test)
- 두 독립 샘플 간의 중앙값 차이를 검정합니다.
- 절차
    1. **데이터 결합**: 두 샘플을 결합하여 전체 데이터 세트를 만듭니다.
    2. **순위 매기기**: 전체 데이터에 순위를 매깁니다.
    3. **순위 합계 계산**: 각 샘플의 순위 합계를 계산합니다.
    4. **검정 통계량 계산**: 순위 합계를 기반으로 U 통계량을 계산합니다.
    5. **유의성 검정**: 유의성 수준에 따라 검정 통계량을 검정합니다.

#### 9) 카이제곱 검정 (Chi-Square Test)
- 범주형 데이터의 빈도 분포를 분석하거나 두 변수 간의 독립성을 평가합니다.
- 절차
    1. **관찰 빈도 계산**: 실제 관찰된 빈도를 계산합니다.
    2. **기대 빈도 계산**: 기대되는 빈도를 계산합니다.
    3. **카이제곱 통계량 계산**: 관찰 빈도와 기대 빈도의 차이를 기반으로 카이제곱 통계량을 계산합니다.
    4. **유의성 검정**: 유의성 수준에 따라 카이제곱 통계량을 검정합니다.
