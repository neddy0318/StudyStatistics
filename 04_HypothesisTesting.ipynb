# 가설 검정 Hypothesis Testing

## 1. 개요 : 가설검정이란?

- 통계적 검정 : 내가 세운 가설이 통계적으로 유의한가를 확인하는 작업을 의미합니다.
- 증명된바 없는 주장 혹은 가설을 표본 통게량에 입각해서 그 진위 여부를 판단, 증명, 검정하는 통계적 추론 방식입니다.
- 표본에서 얻은 사실을 근거로 하여 모집단에 대한 가설이 맞는지 통게적으로 검정하는 분석 방법입니다.

## 2. 구성 및 순서

#### 1. 귀무가설과 대립가설 설정

- 1)귀무가설 Null Hypthesis
    - 직접 검정 대상이 되는 가설로, 표본을 관찰하고 '이 자료들은 이러할 것이다'라고 세운 가설입니다.
    - 증명된바 없는 가설이자 동시에 진실일 가능성이 높아 처음부터 기각(reject) 하는 것이 목표입니다.
    - 가설 검정은 일단 귀무가설이 일단 옳다는 가정에서 시작합니다.
    - 보통 H0으로 표기합니다.
- 2)대립가설 Alternative Hypothesis
    - 귀무가설에 대립되는 가설로, 귀무가설이 기각될 때 받아들여지는 가설입니다.
    - 입증하고자 하는 가설로, 채택(accept)이 목표입니다.
    - 보통 H1으로 표기합니다.

#### 2. 검정 통계량 결정 및 계산
- 1)검정통계량 Test Statistic
    - 검정통계량은 귀무가설이 참이라는 가정 아래 얻은 통계량이자, 귀무가설을 기각할지를 결정하기 위해 계산된 통계량입니다.
    - 표본 데이터를 기반으로 특정 통계적 가설의 타당성을 평가하는 수치입니다.
- 2)검정통계량 계산
    - 표본 데이터를 바탕으로 검정통계량을 계산합니다. 이 통계량은 표본의 평균, 비율, 분산 등을 이용해 계산되며, 가설 검정의 종류에 따라 다릅니다.
    - 예를 들어, 평균 비교 시 t-검정에서는 t-값, 비율 비교 시 z-검정에서는 z-값이 사용됩니다.
- 3)검정통계량의 분포:
    - 검정통계량은 귀무가설이 참일 때 특정 분포를 따르게 됩니다. 예를 들어, t-검정에서는 t-분포, z-검정에서는 표준정규분포를 따릅니다.

- #검정통계량 예시
- (1)t-검정에서의 검정통계량
    - **t-검정 통계량 공식:** t-검정에서 검정통계량(t-값)을 계산하는 과정은 다음과 같습니다:
    - 계산된 t-값은 t-분포에서 어느 위치에 해당하는지 결정됩니다. 이 값이 극단적일수록 (즉, t-값이 크거나 작을수록) 귀무가설이 기각될 가능성이 높아집니다.

$$
t = \frac{\bar{X} - \mu_0}{\frac{s}{\sqrt{n}}}
$$

        - X는 표본의 평균,
        - μ0는 귀무가설에서 주장하는 모집단 평균,
        - s는 표본의 표준편차,
        - n은 표본의 크기입니다.

- (2)z-검정 (Z-test): 
    - z-검정은 모집단의 분산을 알고 있거나, 표본 크기가 충분히 클 때(보통 30 이상) 평균에 대한 가설을 검정하는 방법입니다. 이 검정에서는 표본 평균이 모집단 평균과 얼마나 다른지를 표준정규분포(평균 0, 분산 1)를 사용하여 평가합니다.
    - z-값이 표준정규분포의 양극단에 위치할수록(즉, z-값이 크거나 작을수록) 귀무가설을 기각할 가능성이 높습니다.

    $$
    z = \frac{\bar{X} - \mu_0}{\frac{\sigma}{\sqrt{n}}}
    $$
        - σ는 모집단의 표준편차,
        - 나머지 기호는 t-검정과 동일합니다.


- (3)카이제곱검정 (Chi-Square Test):
    - 카이제곱검정은 범주형 자료에서 관찰된 빈도가 기대되는 빈도와 유의미하게 다른지 검정하는 방법입니다. 주로 두 변수 간의 독립성 검정이나 데이터가 특정 분포에 잘 적합하는지를 검토하는 적합성 검정에 사용됩니다.
    - 카이제곱 통계량이 크면, 관찰된 빈도와 기대되는 빈도 간 차이가 크다는 의미로, 이는 귀무가설을 기각할 가능성을 증가시킵니다. 계산된 χ2 값은 카이제곱 분포와 비교하여 검정합니다.
    $$
    \chi^2 = \sum \frac{(O_i - E_i)^2}{E_i}
    $$
        - Oi는 관찰된 빈도,
        - Ei는 기대되는 빈도입니다.


- (4)F-검정 (F-test):
    - F-검정은 두 개 이상의 집단 간의 분산을 비교하여 집단 간 변동이 유의미한지를 검토하는 방법입니다. 주로 분산분석(ANOVA)에서 사용되며, 여러 그룹의 평균이 동일한지를 검정할 때 사용됩니다.
    - 계산된 F-값이 크면, 집단 간 차이가 크다는 의미로, 이는 집단 간 평균의 차이가 유의미함을 나타낼 수 있습니다. F-값은 F-분포와 비교하여 검정됩니다.
    $$
    F = \frac{\text{집단 간 분산}}{\text{집단 내 분산}}
    $$

#### 2. p-value와 유의수준 설정에 따른 귀무가설 기각 혹은 채택
- 1)p-value (유의확률)
    - p-value는 귀무가설이 참일 때, 관찰된 데이터 또는 더 극단적인 데이터가 나올 확률을 의미합니다. 즉, 표본 데이터가 귀무가설을 얼마나 지지하는지를 나타내는 지표입니다.
    - 해석
        - p-value가 작을수록, 관찰된 데이터가 귀무가설과 일치할 가능성이 낮다는 것을 의미하며, 따라서 귀무가설을 기각할 가능성이 높아집니다.
        - 예를 들어, p-value가 0.03이라면, 귀무가설이 참일 때, 관찰된 결과가 나올 확률이 3%라는 의미입니다.

- 2)유의수준 Significance Level, α
    - 유의수준은 연구자가 1종 오류를 허용할 최대 확률을 의미하며, 귀무가설을 기각하는 기준이 되는 임계값입니다. 일반적으로 0.05(5%)로 설정됩니다.
    - 해석
        - 유의수준이 0.05라면, 연구자는 5%의 확률로 귀무가설이 참인데도 이를 기각하는 오류(1종 오류)를 허용하겠다는 의미입니다.
        - p-value와 비교하여, p-value가 유의수준보다 작으면 귀무가설을 기각하고, 크면 기각하지 않습니다.

- 3)p-value와 유의수준의 관계
    - **p-value ≤ 유의수준**: 귀무가설을 기각합니다. 이는 관찰된 데이터가 귀무가설 하에서 일어날 가능성이 매우 낮기 때문에, 귀무가설이 틀렸다고 결론내립니다.
    - **p-value > 유의수준**: 귀무가설을 기각하지 않습니다. 이는 관찰된 데이터가 귀무가설 하에서 발생할 가능성이 충분히 크다는 의미입니다.

- #예시 : 유의수준 α = 0.05
  - p-value = 0.03인 경우: p-value가 유의수준보다 작으므로 귀무가설을 기각합니다.
  - p-value = 0.07인 경우: p-value가 유의수준보다 크므로 귀무가설을 기각하지 않습니다.



## 3. 신뢰수준과 오류

#### 1)신뢰수준 (Confidence Level)
- 신뢰수준은 구간추정에서 신뢰구간이 실제 모수를 포함할 확률을 의미합니다. 예를 들어, 95% 신뢰수준은 동일한 방법으로 표본을 반복적으로 추출했을 때, 95%의 신뢰구간이 모수를 포함할 것이라는 의미입니다. 
  - 신뢰수준이 95%인 경우, 신뢰구간이 모수를 포함할 확률이 95%라는 것을 의미합니다. 즉, 신뢰구간이 100번 중 95번은 모수를 포함할 것이라고 예측합니다.
  - 신뢰수준이 높을수록 신뢰구간이 넓어지고, 신뢰수준이 낮을수록 신뢰구간이 좁아집니다.

#### 2)오류 유형
- **1종 오류 (Type I Error)**:
  - **정의**: 1종 오류는 귀무가설이 참일 때 이를 잘못 기각하는 오류입니다. 즉, 실제로는 차이나 효과가 없는데도 불구하고 차이나 효과가 있다고 결론 내리는 오류입니다.
  - **확률**: 1종 오류의 확률은 유의수준 (α)으로 설정됩니다. 예를 들어, 유의수준이 0.05이면 1종 오류의 확률은 5%입니다.

- **2종 오류 (Type II Error)**:
  - **정의**: 2종 오류는 귀무가설이 거짓일 때 이를 잘못 기각하지 않는 오류입니다. 즉, 실제로 차이나 효과가 있는데도 불구하고 차이나 효과가 없다고 결론 내리는 오류입니다.
  - **확률**: 2종 오류의 확률은 β로 표시되며, 검정통계의 힘(검정력)과 관련이 있습니다. 검정력은 1 - β로 정의되며, 실제로 대립가설이 맞을 때 이를 올바르게 발견할 확률입니다.

### 3)신뢰수준과 오류의 관계
- **신뢰수준**: 신뢰수준이 높으면 신뢰구간이 넓어져 모수를 포함할 가능성이 높아집니다. 하지만, 신뢰수준이 높아지면 1종 오류(기각 오류)의 가능성은 줄어들지만, 2종 오류(놓치는 오류)의 가능성은 증가할 수 있습니다.
- **유의수준(α)과 1종 오류**: 유의수준은 1종 오류의 확률을 결정하며, 신뢰수준이 높을수록 유의수준이 낮아져 1종 오류의 확률이 줄어듭니다.
- **검정력과 2종 오류**: 검정력은 2종 오류의 확률(β)과 반비례합니다. 즉, 검정력이 높을수록 2종 오류의 확률은 낮아집니다.

### #예시
- **신뢰수준 95%**: 
  - 신뢰구간을 계산했을 때, 이 신뢰구간이 실제 모수를 포함할 확률이 95%입니다.
- **유의수준 α = 0.05**:
  - 1종 오류의 확률이 5%로 설정됩니다.
- **검정력 80%**:
  - 검정력 80%는 실제로 차이가 있을 때, 80%의 확률로 이를 올바르게 발견할 수 있음을 의미합니다. 즉, 2종 오류의 확률 β는 20%입니다.
