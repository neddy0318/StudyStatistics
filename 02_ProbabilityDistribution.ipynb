# 확률분포 Probability Distribution

## 1. 확률분포 개요

- 확률분포는 확률변수가 특정한 값을 가질 확률입니다.

## 2. 이산확률분포

- 이산확률분포는 이산확률 변수가 갖는 확률 분포입니다. 

#### 1)베르누이 분포
- `베르누이 시행` : 동전던지기와 같이 임의의 결과가 '성공' 또는 '실패'의 두 가지 중 하나인 실험을 의미합니다. 이 때, 각 `시행(결과)는 독립적이어서 서로에게 영향을 미치지 않습니다`.
- 베르누이 시행에서 어떤 확률변수 X의 시행 결과가 성공이면 1, 실패면 0의 값을 갖는다고 할 때 이 x를 바로 베르누이 확률변수라고 하며, 이의 분포를 베르누이 분포라고 합니다.
- 이산분포에서 가장 기초적인 분포입니다.

#### 2)이항분포
- n번의 베르누이 시행에서 K번 성공할 확률의 분포입니다.
- 즉, n=1일 때, 이항분포는 베르누이 분포입니다.

#### 3)다항분포

- 이항분포의 확장으로, 이항분포의 각 시행에서 확률변수가 가질 수 있는 경우의 수가 성공/실패 두 가지였다면, 다항분포는 경우의 수가 k가지입니다.

#### 3)포아송분포

**(1)개요**
- 포아송분포는 단위 시간, 단위 공간에서 어떤 사건이 몇 번 발생할 것인가를 표현하는 이산 확률 분포입니다.
- 이해를 위해 이항분포와 비교해보도록 하겠습니다.
    - 이항분포 : n번 중 성공확률이 p일 때 성공 횟수를 확률변수로 하는 분포
    - 포아송분포 : 일정 단위에서 평균 성공 수가 m일 때 성공 횟수를 확률변수로 하는 분포(일정 단위 = 시간, 공간 등)
- 예를 들면 어떤 공장에서 10시간(일정 단위)마다 평균 2개의 불량품(평균 성공 수 m)이 발생한다면 불량품이 하나도 없는 경우부터 수십, 수백개가 발생할 확률까지 성공 횟수에 따른 확률을 다루는 분포입니다.
- 이와 같이 빈도를 적용할 수 있는 데이터에 전부 적용할 수 있기에 일상에서 많이 사용합니다(한 시간 동안 상점을 방문한 고객 수, 한 시간 동안 사무실에 걸려온 전화 수 등).
- 'n번 중 성공 횟수'의 분포인 이항분포와 매우 밀접한 관련이 있는데, 이항분포의 평균이 np이고 이는 곧 '평균 성공 수'의 관점으로 바라볼 수 있기 때문입니다.
- 평균과 분산이 같습니다.

**(2)전제조건**
- 독립성
    - 특정 단위 시간 또는 공간에서 발생한 결과는 중복되지 않은 다른 시간 혹은 공간에서 발생한 결과와 서로 독립되어야 합니다.
    - ex)09~10시에 A상점에 방문한 고객 수와 B상점에 방문한 고객 수는 독립, A상점에 10~11시에 방문한 고객수와도 독립
- 일정성
    - 특정 단위 시간 또는 공간에서 발생한 확률은 해당 시간/공간의 크기에 비례하고 외부의 영향을 받지 않습니다.
    - 예를 들어, 어떤 사건이 1분에 평균 2회 발생하면 3분에는 평균 6회 발생합니다.
- 비집락성
    - 두 개 이상의 결과가 동시에 발생할 확률은 0입니다.

**(3)지수분포**
- 포아송 분포에 의한 사건이 발생할 때까지 걸리는 시간의 확률분포입니다.

## 3. 연속확률분포

- 연속확률분포는 발생 가능한 값이 무한대이기 때문에 이산확률분포와 달리 발생 가능한 모든 값에 확률을 대응하고 나열하여 표현하기 어렵습니다.
- 그렇기에 주어진 구간에서 확률이 어떻게 분포하는지 판단하고, 어느 구간의 확률이 작고 큰지를 나타낼 수 잇는 함수를 사용합니다.
- 연속확률변수 X의 확률분포는 확률의 밀도를 나타내는 확률밀도함수에 의해 결정됩니다.
- 연속활률분포에서 엄밀하게 확률의 값을 정의하기 위해서는 '특정 구간'이 주어져야 합니다.

#### 1)정규분포

- 자연의 수많은 확률을 반영하는 확률밀도함수로, 평균과 분산에 의해 그 분포가 확정됩니다.
- 확률변수 X가 평균 μ, 표준편차 σ인 정규분포를 따를 때 주로 N(μ,σ^2)로 표현하게 됩니다. 표준편차 σ가 커질수록 정규분포의 모형은 옆으로 퍼지게 됩니다.
- 표준정규분포 : 평균이 0이고 표준편차가 1인 정규분포를 의미합니다. 표준정규분포를 가지는 확률변수를 Z라고 표현했을 때 확률변슈 Z는 0을 중심으로 대칭인 분포를 가지게 됩니다.

- 표준정규분포표 : 표준정규분포를 표로 나타낸 표입니다. 이를 활용하기 위해서는 표준정규확률변수 z를 구해야 합니다.

## 포아송 분포 코드

```python
#기대값이 10일 때

import numpy as np
import matplotlib.pyplot as plt
from math import factorial, exp

def pois_dist(n, lamb):
    pd = (lamb ** n) * exp(-lamb) / factorial(n)
    return pd


x= np.arange(40)
pd1 = np.array([pois_dist(n, 10) for n in range(40)])
plt.ylim(0, 0.15)
plt.text(33.5, 0.14, 'lamb = 10')
plt.bar(x, pd1, color = 'lightcoral')
plt.show()
```

```python
#기대값의 변화

x = np.arange(40)
plt.subplot(3, 1, 1)
pd1 = np.array([pois_dist(n, 10) for n in range(40)])
plt.ylim(0, 0.15)
plt.text(33.5, 0.12, 'lamb = 10')
plt.bar(x, pd1, color='lightcoral')

plt.subplot(3, 1, 2)
pd2 = np.array([pois_dist(n, 15) for n in range(40)])
plt.ylim(0, 0.15)
plt.text(33.5, 0.12, 'lamb = 15')
plt.bar(x, pd2, color='mediumaquamarine')

plt.subplot(3, 1, 3)
pd3 = np.array([pois_dist(n, 20) for n in range(40)])
plt.ylim(0, 0.15)
plt.text(33.5, 0.12, 'lamb = 20')
plt.bar(x, pd3, color='royalblue')

plt.show()
```

## 정규분포 코드

```python
from scipy.stats import norm
import pandas as pd

mu = 10
std = 1
norm_ = norm(mu, std) #평균 10, 표준편차 1인 정규분포 생성

# x 값 생성: 정규분포의 1% 백분위수와 99% 백분위수 사이의 x 값을 1000개의 점으로 나누어 배열을 생성합니다.
x = np.linspace(norm_.ppf(0.01), norm_.ppf(0.99), 1000)

# x 값에 대한 정규분포의 확률밀도함수(PDF)를 계산하여 그래프를 그립니다.
# 'r-'는 빨간색 실선을 의미하고, lw=5는 선의 두께를 5로 설정합니다.
# alpha=0.6은 선의 투명도를 0.6으로 설정합니다.
# label='norm pdf'는 범례에서 사용할 레이블을 설정합니다.
plt.plot(x, norm_.pdf(x), 'r-', lw=5, alpha=0.6, label='norm pdf')

# 그래프에 레이블을 붙여 범례를 생성합니다.
plt.legend()

# 생성된 그래프를 화면에 표시합니다.
plt.show()
```

#### ppf (Percent-Point Function)
- 정의: ppf는 Percent-Point Function의 약자로, 역누적분포함수 (inverse cumulative distribution function) 를 의미합니다. 주어진 확률(퍼센트)에 해당하는 분포의 값을 반환합니다.

- 용도: 확률분포에서 특정 백분위수(또는 확률)에 해당하는 값을 찾을 때 사용됩니다. 예를 들어, 정규분포에서 1% 백분위수에 해당하는 x 값을 알고 싶을 때 ppf를 사용합니다.


#### linspace (Linear Space)
- 정의: linspace는 선형 구간 생성 함수로, 두 지점 사이를 일정한 간격으로 나눈 값을 생성합니다. 이 함수는 주어진 시작점과 끝점 사이를 균일한 간격으로 나눈 값을 포함하는 배열을 생성합니다.

- 용도: 연속적인 데이터를 생성하거나 그래프를 그릴 때 특정 범위의 값들을 균등하게 나누어 생성할 때 사용됩니다. 예를 들어, x축의 구간을 정해놓고 그 구간을 일정한 간격으로 나눈 x 값을 생성할 때 사용됩니다.

```python
data_norm = norm_.rvs(1000) ## 평균 10, 표준편차 1의 정규분포를 따르는 데이터 생성
plt.hist(data_norm, density =True, histtype = "stepfilled", alpha=0.2, label = "norm_hist")
plt.legend()
plt.show()
```

```python
## 동시에 시각화 표현
fig, ax = plt.subplots(1, 1)
ax.plot(x, norm_.pdf(x),'r-', lw=5, alpha=0.6, label='norm pdf')
ax.hist(data_norm, density =True, histtype = "stepfilled", alpha=0.2, label = "simulation")
ax.legend(loc="upper right")
plt.show()
```
