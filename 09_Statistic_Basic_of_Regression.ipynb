# 회귀분석의 통계적 기초

## 1. 회귀분석 개요
1)개요
- 종속 변수와 하나 이상의 독립 변수 간의 관계를 모델링하고 분석하는 통계적 방법입니다.
- 독립변수 x에 대응하는 종속변수 y와 가장 비슷한 값을 출력하는 함수 f(x)를 찾는 과정입니다.
- 변수 간 관계를 이해하고, 예측을 수행하며, 변수들이 결과에 미치는 영향을 추정하는 데 사용됩니다.

2)기본 개념
- 종속 변수(Dependent Variable, Y): 예측하고자 하는 목표 변수입니다. 이 변수는 독립 변수들에 의해 설명되거나 영향을 받는다고 가정합니다.
- 독립 변수(Independent Variable, X): 종속 변수에 영향을 미치는 변수들입니다. 하나의 독립 변수를 사용하는 경우를 단순회귀(Simple Regression), 여러 개의 독립 변수를 사용하는 경우를 다중회귀(Multiple Regression)라고 합니다.
- 회귀모형(Regression Model): 종속변수와 독립 변수 간 관계를 설명하는 수학적 모형입니다.

## 2. 종류

#### 1)단순선형회귀
$$
Y = \beta_0 + \beta_1 X + \epsilon
$$

- Y: 종속 변수, 반응 변수
- X: 독립 변수, 설명 변수
- β₀: 절편, intercept
- β₁: 기울기, slope

이 식에서 β₀는 X가 0일 때 Y의 예상값을 나타내며, β₁는 X가 한 단위 증가할 때 Y가 변화하는 정도를 나타냅니다.
#### 2)다중회귀

다중회귀분석은 여러 개의 독립 변수를 사용하여 종속 변수를 설명하는 모형입니다. 이 경우 회귀모형은 다음과 같이 확장됩니다.

$$
Y = \beta_0 + \beta_1 X_1 + \beta_2 X_2 + \dots + \beta_p X_p
$$

- X₁, X₂, ..., Xₚ: 여러 개의 독립 변수


## 3. 주요 가정
1)선형성(Linearity):
- 독립변수와 종속변수 간 선형적인 관계가 선형적이어야 합니다.
- 종속변수와 독립변수 간 직선의 형태를 띤 관계가 없을 경우 회귀분석은 불가능합니다.

2)독립성(Independency): 
- 다중회귀분석에서 여려 독립변수 간 특정한 관계가 없어야합니다.
- 독립성 가정이 위배될 경우 다중공선성(Multicollinearity) 문제가 발생하고, 회귀 계수의 분산이 비정상적으로 커져 계수 추정이 어려워집니다.

3)등분산성(Homoskedasticity):
- 오차의 분산이 모든 수준의 독립 변수에 대해 일정해야 합니다.
- 즉, 오차가 특정 패턴을 보여서는 안 됩니다.

4)정규성
- 오차가 정규분포의 형태를 띠어야합니다.


## 4. 회귀 계수 추정 : 최소제곱법
- 회귀분석에서는 주어진 데이터를 사용해 회귀 계수(β₀와 β₁)를 추정합니다.
- 가장 많이 사용되는 방법은 **최소제곱법**입니다.
- 최소제곱법은 잔차 제곱합 RSS를 최소화하는 β₀와 β₁를 찾는 방법입니다.

$$
\text{RSS} = \sum_{i=1}^{n} (Y_i - \hat{Y_i})^2
$$

- RSS: 잔차 제곱합, Residual Sum of Squares
- Yᵢ: 관측된 종속 변수 값
- Ŷᵢ: 예측된 종속 변수 값


## 5. 회귀분석 관련 통계 지표들

#### 1)SST (Total Sum of Squares, 총 제곱합)
- SST는 종속 변수 Y의 전체 변동성을 나타내는 지표입니다.
- 이는 데이터의 분산을 나타내며, 회귀모형이 얼마나 잘 데이터를 설명할 수 있는지를 판단하는 기준이 됩니다.

  $$
  \text{SST} = \sum_{i=1}^{n} (Y_i - \bar{Y})^2
  $$
  (Y_i는 관측된 종속 변수의 값, Ȳ는 종속 변수의 평균)

- **설명**: SST는 모든 관측값 Y_i가 평균 Ȳ로부터 얼마나 떨어져 있는지를 제곱합한 값입니다. 즉, SST는 종속 변수 Y의 총 변동성을 나타내며, 회귀모형이 이 변동성을 얼마나 설명할 수 있는지를 나타냅니다.

#### 2)SSR (Sum of Squared Residuals, 잔차 제곱합)
- SSR은 회귀모형이 설명하지 못한 변동성, 즉 잔차의 변동성을 나타냅니다.
- 잔차는 실제 관측값과 회귀모형이 예측한 값의 차이를 말합니다.

  $$
  \text{SSR} = \sum_{i=1}^{n} (Y_i - \hat{Y_i})^2
  $$
  (Ŷ_i는 회귀모형이 예측한 종속 변수의 값)

- **설명**: SSR은 회귀모형이 설명하지 못한 부분, 즉 데이터의 실제값 Y_i와 모형이 예측한 값 Ŷ_i 간의 차이에서 발생하는 변동성을 나타냅니다. 이 값이 작을수록 회귀모형이 데이터를 잘 설명하고 있다는 의미입니다.

#### 3)MSR (Mean Square Regression, 회귀 제곱평균)
- MSR은 회귀모형이 설명한 변동성을 나타내는 SSR을 독립 변수의 개수로 나누어 평균화한 값입니다.
- 이는 회귀모형의 설명력을 평가하는 지표로 사용됩니다.

  $$
  \text{MSR} = \frac{\text{SSR}}{p}
  $$
  (여기서 p는 독립 변수의 개수(또는 자유도))

- **설명**: MSR은 SSR을 독립 변수의 수로 나눈 값으로, 각 독립 변수가 종속 변수에 얼마나 기여하는지를 나타냅니다. MSR은 회귀모형의 설명력을 독립 변수의 수에 대해 조정한 값으로, F-검정에서 MSR을 잔차 제곱평균(MSE)으로 나누어 회귀모형의 유의성을 평가할 때 사용됩니다.



## 6. 통계적 지표를 활용한 회귀모형의 적합성 평가

#### 1)결정계수 R²
- 회귀모형이 종속 변수의 변동성을 얼마나 설명하는지 나타내는 척도입니다.

$$
R^2 = 1 - \frac{\text{SSR}}{\text{SST}}
$$

- SSR: 설명되지 않은 변동, Sum of Squared Residuals
- SST: 총 변동, Total Sum of Squares

#### 2)F-검정
- 회귀모형 전체의 유의성을 검정하는 방법입니다. 모형이 통계적으로 유의미한지 여부를 판단하는 데 사용됩니다.

$$
F = \frac{\text{MSR}}{\text{MSE}}
$$

- MSR: 회귀의 평균제곱, Mean Square Regression
- MSE: 잔차의 평균제곱, Mean Square Error

#### 3)t-검정

- 회귀분석에서 t-검정은 각 독립 변수의 회귀계수가 유의미한지 검정하는 데 사용됩니다.
- 이를 통해 특정 독립 변수가 종속 변수에 유의미한 영향을 미치는지 확인할 수 있습니다.
- 각 회귀계수 βᵢ에 대해 다음과 같은 가설을 검정합니다:
    - **귀무가설 H₀**: βᵢ = 0 (독립 변수 Xᵢ가 종속 변수 Y에 유의미한 영향을 미치지 않는다)
    - **대립가설 H₁**: βᵢ ≠ 0 (독립 변수 Xᵢ가 종속 변수 Y에 유의미한 영향을 미친다)

- t-검정 통계량은 다음과 같이 계산됩니다:

$$
t = \frac{\hat{\beta_i}}{SE(\hat{\beta_i})}
$$

- **βᵢ**: 독립 변수 **Xᵢ**의 추정된 회귀계수
- **SE(βᵢ)**: 회귀계수의 표준 오차<br/><br/>

- 이 t-값을 표준 정규분포에서 특정 임계값과 비교하여 귀무가설을 기각할지 여부를 결정합니다.
- p-값이 미리 설정된 유의수준(예: 0.05)보다 작으면, 귀무가설을 기각하고 Xᵢ가 Y에 유의미한 영향을 미친다고 결론지을 수 있습니다.


#### 4)잔차분석
- 잔차(실제값과 예측값 간의 차이)를 분석하여 모형의 가정이 잘 충족되는지, 이상치가 있는지 등을 평가합니다.

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
#예제 데이터 생성
np.random.seed(0)
X = np.random.rand(100, 2)
y = 2 + 3 * X[:, 0] + 4 * X[:, 1] + np.random.randn(100)
```

```python
# 데이터 프레임으로 변환
data = pd.DataFrame(X, columns=['X1', 'X2'])
data['Y'] = y
```

```python
data
```

```python
# 회귀 모델 적합
model = smf.ols('Y ~ X1 + X2', data=data).fit()

# 회귀 분석 결과 요약
print(model.summary())

# 결정계수 R² 계산
r_squared = model.rsquared
print(f"결정계수 R²: {r_squared:.4f}")

# F-검정
f_value = model.fvalue
f_pvalue = model.f_pvalue
print(f"F-검정 통계량: {f_value:.4f}, p-값: {f_pvalue:.4f}")

# t-검정
for i in range(1, len(model.params)):
    param_name = model.params.index[i]
    t_value = model.tvalues[i]
    p_value = model.pvalues[i]
    print(f"회귀계수 {param_name}의 t-검정 통계량: {t_value:.4f}, p-값: {p_value:.4f}")

# 잔차 분석
residuals = model.resid
fitted_values = model.fittedvalues

plt.figure(figsize=(12, 6))

# 잔차 플롯
plt.subplot(1, 2, 1)
plt.scatter(fitted_values, residuals)
plt.axhline(0, color='red', linestyle='--')
plt.xlabel('Fitted Values')
plt.ylabel('Residuals')
plt.title('Residual Plot')

# Q-Q 플롯
plt.subplot(1, 2, 2)
sm.qqplot(residuals, line='s', ax=plt.gca())
plt.title('Q-Q plot')

plt.tight_layout()
plt.show()
```
